```
import pandas as pd
import requests
import json

url = 'https://www.aivivn.com/graphql'
payload = {"operationName":"ContestTeamQuery","variables":{"id":"10","skip":10},"query":"query ContestTeamQuery($id: ID!, $skip: Int) {\n  contest(id: $id) {\n    id\n    endDate\n    teams(skip: $skip) {\n      totalCount\n      nodes {\n        id\n        name\n        rank\n        publicRank\n        lastWeekRank\n        publicScore\n        privateScore\n        numSubmissions\n        lastSubmissionAt\n        members {\n          id\n          username\n          user {\n            avatar\n            __typename\n          }\n          __typename\n        }\n        __typename\n      }\n      __typename\n    }\n    __typename\n  }\n}\n"}
df = pd.DataFrame()
for i in range(0,85):
  payload = {"operationName":"ContestTeamQuery","variables":{"id":"10","skip":i*10},"query":"query ContestTeamQuery($id: ID!, $skip: Int) {\n  contest(id: $id) {\n    id\n    endDate\n    teams(skip: $skip) {\n      totalCount\n      nodes {\n        id\n        name\n        rank\n        publicRank\n        lastWeekRank\n        publicScore\n        privateScore\n        numSubmissions\n        lastSubmissionAt\n        members {\n          id\n          username\n          user {\n            avatar\n            __typename\n          }\n          __typename\n        }\n        __typename\n      }\n      __typename\n    }\n    __typename\n  }\n}\n"}
  r = requests.post(url, json = payload)
  data = json.loads(r.text)
  df = df.append(pd.DataFrame(data['data']['contest']['teams']['nodes']))
```
